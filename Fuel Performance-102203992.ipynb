<h2>Maximum R2 score is by Adaboost Regressor</h2>

```python
Model = create_model('ada')
```

<h3>Making predictions on unseen data</h3>

```python
newdata=data1.iloc[1:10]
newdata
```

```python
pred=predict_model(Model, data = newdata)
pred
```

<h2>Plot : Actual v/s Predicted</h2>

```python
import matplotlib.pyplot as plt
predicted = pred.iloc[:,-1]     # Last column
actual = pred.iloc[:,-2]        # 2nd last column
plt.scatter(actual, predicted)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Actul Vs Predicted')
plt.show()
```
